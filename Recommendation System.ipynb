```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#EDA
#renaming the columns
```

```python
df=df.rename(columns={"User.ID":"user_id","Book.Title":"book_title","Book.Rating":"book_rating"})
```

```python
#creating pivot table for easy understanding
```

```python
user_df=df.pivot_table(index="user_id",columns="book_title",values="book_rating", aggfunc='mean')
```

```python
user_df
```

```python
user_df.head()
```

```python
#fillling the null values with 0
```

```python
user_df.fillna("0",inplace=True)
```

```python
user_df
```

```python
#calculating cosine similarity betwwen the users
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_simi= 1- pairwise_distances(user_df.values,metric="cosine")
user_simi
```

```python
#storing the similarity in a dataframe
```

```python
user_simi_df=pd.DataFrame(user_simi)
```

```python
user_simi_df
```

```python
df.user_id.unique()
```

```python
#setting index and column names to userid
```

```python
user_simi_df.index=df.user_id.unique()
user_simi_df.columns=df.user_id.unique()
```

```python
user_simi_df.iloc[0:20,0:20]
```

```python
#filling the diagonal values to 0
```

```python
np.fill_diagonal(user_simi,0)
user_simi_df.iloc[0:20,0:20]
```

```python
user_simi_df.idxmax(axis=1)[0:20]
```
